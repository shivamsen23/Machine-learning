```python
import pandas as pd
import seaborn as sns
```

```python
tips= sns.load_dataset('tips');
tips.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
iris= sns.load_dataset('iris')
```

**1.Scatterplot(Numerical- Numerical)**

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'],hue=tips['sex'])
```

**2.Bar Plot(Numerical-Categorical)**

```python
sns.barplot(x=tips['sex'],y=tips['total_bill'],hue=tips['smoker'])
```

**3. Distplot(Numerical-categorical)**

**4. HeatMap(Categorical-categorical)**

```python
sns.heatmap(pd.crosstab(tips['sex'],tips['smoker']))
```

```python
tips.groupby('total_bill').mean()['smoker']
```

**6. PairPlot**

```python
sns.pairplot(iris,hue='species');
```

*** 7.LinePlot(Numerical-Numerical)***

```python
new= flights.groupby('year')['passengers'].sum().reset_index()
sns.lineplot(x=new['year'], y=new['passengers'])
```
